```python
class BankAccount:
    def __init__(self, account_number, name, initial_deposit=0):
        self.account_number = account_number
        self.name = name
        self.balance = initial_deposit
        self.transactions = [("Initial Deposit", initial_deposit)]

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(("Deposit", amount))
            print(f"Deposited {amount} to account {self.account_number}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            self.transactions.append(("Withdrawal", amount))
            print(f"Withdrew {amount} from account {self.account_number}")
        else:
            print("Insufficient balance or invalid amount.")

    def check_balance(self):
        print(f"Account {self.account_number} balance: {self.balance}")
        return self.balance

    def transfer(self, recipient, amount):
        if 0 < amount <= self.balance:
            self.withdraw(amount)
            recipient.deposit(amount)
            print(f"Transferred {amount} from account {self.account_number} to account {recipient.account_number}")
        else:
            print("Insufficient balance or invalid amount.")

    def view_transaction_history(self):
        print(f"Transaction history for account {self.account_number}:")
        for transaction, amount in self.transactions:
            print(f"{transaction}: {amount}")

class Bank:
    def __init__(self):
        self.accounts = {}

    def open_account(self, name, initial_deposit):
        account_number = len(self.accounts) + 1
        account = BankAccount(account_number, name, initial_deposit)
        self.accounts[account_number] = account
        print(f"Account {account_number} opened for {name} with initial deposit {initial_deposit}.")
        return account

    def view_total_deposits(self):
        total_deposits = sum(account.balance for account in self.accounts.values())
        print(f"Total deposits in the bank: {total_deposits}")
        return total_deposits

    def total_accounts(self):
        print(f"Total number of accounts: {len(self.accounts)}")
        return len(self.accounts)

# Example usage
if __name__ == "__main__":
    bank = Bank()

# Open new accounts
    account1 = bank.open_account("Alice", 1000)
    account2 = bank.open_account("Bob", 500)

# Perform transactions
    account1.deposit(500)
    account1.withdraw(300)
    account1.check_balance()
    
    account1.transfer(account2, 200)

# View transaction history
    account1.view_transaction_history()
    account2.view_transaction_history()

# Admin actions
    bank.view_total_deposits()
    bank.total_accounts()
```
